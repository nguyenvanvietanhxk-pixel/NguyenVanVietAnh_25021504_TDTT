Bài 2

```python
n = int(input())
if n < 2:
    print("Không phải số nguyên tố")
else:
    i = 2
    isPrime = True
    while i * i <= n:
        if n % i == 0:
            isPrime = False
            break
        i += 1
    if isPrime:
        print("Là số nguyên tố")
    else:
        print("Không phải số nguyên tố")
```

```python
a, b = map(int, input().split())
s = 0
while a <= b:
    # kiểm tra x có phải số nguyên tố
    if a >= 2:
        i = 2
        check = True
        while i * i <= a:
            if a % i == 0:
                check = False
                break
            i += 1
        if check:
            s += a
    a += 1

print(s)
```

Bài 10

```python
n = int(input())

def collatz_len(x):
    cnt = 1
    while x != 1:
        x = x // 2 if x % 2 == 0 else 3 * x + 1
        cnt += 1
    return cnt

best_x = 1
best_len = collatz_len(1)

for i in range(2, n + 1):
    L = collatz_len(i)
    if L > best_len:
        best_len = L
        best_x = i

print(best_x, best_len)
```

```python
n = int(input())
cnt = 0

for i in range(2, int(n**0.5) + 1, 2):
    if n % i == 0:
        cnt += 1
        if i != n // i and (n // i) % 2 == 0:
            cnt += 1

if n % 2 == 0:
    cnt += 1

print(cnt)
```
