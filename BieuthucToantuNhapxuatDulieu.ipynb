```python
dtb = float(input("Nhập điểm trung bình: "))
if dtb >= 8:
    print("Xuất sắc")
elif dtb >= 6.5:
    print("Giỏi")
elif dtb >= 5.5:
    print("Khá")
elif dtb >= 4:
    print("Trung bình")
else:
    print("Yếu")
```
