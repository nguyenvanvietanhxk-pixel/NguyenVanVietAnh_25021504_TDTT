```python
def max(a, b):
    if a > b:
        return a
    else:
        return b
a, b = map(int, input("Nhập hai số a và b: ").split())
print("Số lớn hơn là:", max(a, b))
```

Bài 3

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input("Nhập một số nguyên dương: "))
print(is_prime(n))
```

Bài 4

```python
def is_perfect(n):
    if n < 2:
        return False
    
    total = 1 
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            total += i
            other = n // i
            if other != i:   
                total += other
    
    return total == n
n = int(input())
print(is_perfect(n))
```

Bài 5

```python
def find_first_position(lst, k):
    for i in range(len(lst)):  
        if lst[i] == k:         
            return i            
    return -1                   
lst = list(map(int, input().split()))
k = int(input())
pos = find_first_position(lst, k)
print("Vị trí của k:", pos)
```

Bài 6

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
n = int(input())
print("Giai thừa của", n, "là:", factorial(n))
```

Bài 7

```python
def calculator(num1, num2, operat):
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Lỗi: chia cho 0"
        result = num1 / num2
    else:
        return "Toán tử không hợp lệ"

    return round(result, 2)   # làm tròn 2 chữ số thập phân
num1 = float(input("Nhập số thứ nhất: "))
num2 = float(input("Nhập số thứ hai: "))
operat = input("Nhập toán tử (+, -, *, /): ")
print("Kết quả:", calculator(num1, num2, operat))
```

Bài 8

```python
def hamming_distance(x, y):
    xor_value = x ^ y             
    count = 0                      

    while xor_value > 0:
        count += xor_value & 1    
        xor_value >>= 1            
    
    return count
x = int(input("Nhập số nguyên x: "))
y = int(input("Nhập số nguyên y: "))
print("Khoảng cách Hamming giữa", x, "và", y, "là:", hamming_distance(x, y))
```

```python
def hamming_distance(x, y):
    return bin(x ^ y).count('1')
x = int(input("Nhập số nguyên x: "))
y = int(input("Nhập số nguyên y: "))
print("Khoảng cách Hamming giữa", x, "và", y, "là:", hamming_distance(x, y))
```

Bài 9

```python
def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10   
        n //= 10         
    return total
n = int(input("Nhập một số nguyên dương: "))
print("Tổng các chữ số của", n, "là:", sum_of_digits(n))
```

```python
def dang_cau(a, b):
    map_a_to_b = {}
    mapped_chars = set()

    for ch1, ch2 in zip(a, b):
        if ch1 in map_a_to_b:
            if map_a_to_b[ch1] != ch2:
                return False
        else:
            if ch2 in mapped_chars:
                return False
            map_a_to_b[ch1] = ch2
            mapped_chars.add(ch2)

    return True
str1 = input("Nhập chuỗi thứ nhất: ")
str2 = input("Nhập chuỗi thứ hai: ")
print(dang_cau(str1, str2))
```
