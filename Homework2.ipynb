```python
#Cách 1
import math
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))
if a == 0:
    if b == 0:
        print("Vô số nghiệm" if c == 0 else "Vô nghiệm")
    else:
        print("Phương trình có 1 nghiệm:", -c/b)
else:
    delta = b*b - 4*a*c
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép x =", x)
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Phương trình có 2 nghiệm phân biệt:")
        print("x1 =", x1)
        print("x2 =", x2)
```

```python
#Cách 2
import math
def giai_pt_bac2(a, b, c):
    if a == 0:
        if b == 0:
            return "Vô số nghiệm" if c == 0 else "Vô nghiệm"
        return f"1 nghiệm: x = {-c/b}"
    delta = b*b - 4*a*c
    if delta < 0:
        return "Vô nghiệm"
    if delta == 0:
        x = -b / (2*a)
        return f"Nghiệm kép x = {x}"
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    return f"2 nghiệm: x1 = {x1}, x2 = {x2}"
a = float(input("a = "))
b = float(input("b = "))
c = float(input("c = "))
print(giai_pt_bac2(a, b, c))
```

```python
#Cách 1
def giai_he_cramer(a, b, c, d, m, n):
    D  = a*d - b*c
    Dx = m*d - b*n
    Dy = a*n - m*c
    if D == 0:
        return "Hệ vô nghiệm hoặc vô số nghiệm"
    x = Dx / D
    y = Dy / D
    return x, y
a, b, c, d, m, n = map(float, input().split())
print(giai_he_cramer(a, b, c, d, m, n))
```

```python
#Cách 2
def giai_gauss(a, b, c, d, m, n):
    # Hệ: [a b | m]
    #      [c d | n]

    # Khử x ở dòng 2
    he_so = c / a
    c2 = c - he_so * a
    d2 = d - he_so * b
    n2 = n - he_so * m

    y = n2 / d2
    x = (m - b*y) / a
    return x, y
a, b, c, d, m, n = map(float, input().split())
print(giai_gauss(a, b, c, d, m, n))
```
