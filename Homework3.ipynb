Bài 4

```python
h = int(input("Nhập chiều cao h: "))
for i in range(1, h + 1):
    # In khoảng trắng bên trái
    for j in range(h - i):
        print(" ", end="")
    if i == 1:
        # Dòng đầu tiên
        print("*")
    elif i == h:
        # Dòng cuối (đáy)
        for j in range(2*h - 1):
            print("*", end="")
        print()
    else:
        # Các dòng giữa
        print("*", end="")
        for j in range(2*i - 3):
            print(" ", end="")
        print("*")
```

Bài 8

```python
a, b = map(int, input("Nhập hai số nguyên dương a và b: ").split())
if a <= 0 or b <= 0:
    print("a và b phải là số nguyên dương")
else:
    i = 1
    while i <= a and i <= b:
        if a % i == 0 and b % i == 0:
            print(i, end=" ")
        i += 1
```

Bài 9

```python
a = int(input("Nhập số nguyên dương a: "))
if a < 0:
    print("a không phải số chính phương")
else:
    i = 0
    while i * i < a:
        i += 1
    if i * i == a:
        print("a là số chính phương")
    else:
        print("a không phải số chính phương")
```

Bài 10

```python
a = int(input("Nhập số nguyên dương a: "))
if a <= 0:
    print("a không phải số hoàn hảo")
else:
    tong = 0
    i = 1
    while i <= a // 2:
        if a % i == 0:
            tong += i
        i += 1
    if tong == a:
        print("a là số hoàn hảo")
    else:
        print("a không phải số hoàn hảo")
```

Bài 12

```python
n = int(input("Nhập số nguyên dương n: "))
if n <= 0:
    print("n phải là số nguyên dương")
else:
    s = 0
    k = 0
    while s + (k + 1) < n:
        k += 1
        s += k
    print("k =", k)
    print("S(k) =", s)
```

Bài 13

```python
A = float(input("Nhập số A (A > 0): "))
tong = 0.0
n = 0
while tong <= A:
    n += 1
    tong += 1 / n

print("n nhỏ nhất là:", n)
print("Tổng =", tong)
```

Bài 14

```python
x = int(input("Nhập số (nhập -1 để kết thúc): "))
if x == -1:
    print("Chưa nhập số nào")
else:
    max_val = x
    min_val = x
    while True:
        x = int(input("Nhập số (nhập -1 để kết thúc): "))
        if x == -1:
            break
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x
    print("Số lớn nhất:", max_val)
    print("Số nhỏ nhất:", min_val)
```

Bài 16

```python
n = int(input("Nhập n: "))
chan = 0
le = 0
while n > 0:
    cs = n % 10
    if cs % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10
print("Chữ số chẵn:", chan)
print("Chữ số lẻ:", le)
```

Bài 19

```python
A = int(input("Nhập A: "))

f1 = 1
f2 = 1

while f2 <= A:
    f = f1 + f2
    f1 = f2
    f2 = f

print("Số Fibonacci lớn nhất không vượt quá A là:", f1)
```
