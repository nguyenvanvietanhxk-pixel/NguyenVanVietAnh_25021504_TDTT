```python
def lt(n):
    return n > 0 and (n & (n - 1)) == 0
numbers = [int(x) for x in input().split()]
for num in numbers:
    if lt(num):
        print(f"{num} là lũy thừa của 2")
    else:
        print(f"{num} không phải là lũy thừa của 2")
```

```python
#Cách 2
m = int(input("Nhập m: "))
n = int(input("Nhập n: "))
if m % n == 0:
    ket_qua = m // n
else:
    if (m > 0 and n > 0) or (m < 0 and n < 0):
        ket_qua = m // n + 1
    else:
        ket_qua = m // n 
print(f"Kết quả của {m} chia {n} làm tròn lên là {ket_qua}")
```
