```python
nums = list(map(int, input().split()))
result = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0
}
for x in nums:
    if x > 0:
        result['positives'] += 1
    elif x < 0:
        result['negatives'] += 1
    else:
        result['zeros'] += 1
print(result)
```

W6A9

```python
n1 = int(input())
d1 = {}
for _ in range(n1):
    k, v = input().split()
    d1[k] = int(v)
n2 = int(input())
d2 = {}
for _ in range(n2):
    k, v = input().split()
    d2[k] = int(v)
d = {}
for k in d1.keys() | d2.keys():
    d[k] = d1.get(k, 0) + d2.get(k, 0)
print(dict(sorted(d.items())))
```

W6A10

```python
n = list(map(int, input().split()))
k = int(input())
b = []
a = len(n)
for i in range(a):
    for j in range(i + 1, a):
        if n[i] + n[j] == k:
            pair = (n[i], n[j])
            if pair not in b:
                b.append(pair)
b.sort(key=lambda x: x[0])
print(b)
```

W6A12

```python
nums = list(map(int, input().split()))
freq = {}
for x in nums:
    freq[x] = freq.get(x, 0) + 1
mx = max(freq.values())
ans = min(k for k, v in freq.items() if v == mx)
print(ans)
```

```python
pairs = input().split()
k = int(input())
d = {}
for p in pairs:
    key, val = p.split(':')
    d[key] = int(val)
res = {key: val for key, val in d.items() if val > k}
print(res)
```
