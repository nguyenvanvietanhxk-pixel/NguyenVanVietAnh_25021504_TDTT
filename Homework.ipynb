```python
c = input("Nhập vào một ký tự: ")
if len(c) != 1:
    print("Vui lòng chỉ nhập một ký tự duy nhất!")
else:
    if c.isupper():
        print("Ký tự sau khi đổi là:", c.lower())
    elif c.islower():
        print("Ký tự sau khi đổi là:", c.upper())
    else:
        print("Ký tự không phải là chữ cái!")
```

Bài 5

```python
c = input("Nhập vào một chữ cái hoa: ")
if len(c) != 1 or not c.isupper():
    print("Vui lòng nhập một chữ cái hoa (A-Z)!")
else:
    if c == 'A':
        print("Trường hợp đặc biệt: 'A' không có chữ cái thường liền trước.")
    else:
        sau_ki_tu = c.lower()
        truoc_ki_tu = chr(ord(sau_ki_tu) - 1)
        print(f"Chữ cái thường liền trước {sau_ki_tu} là: {truoc_ki_tu}")
```

```python
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f"Diện tích tam giác là: {S:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```
