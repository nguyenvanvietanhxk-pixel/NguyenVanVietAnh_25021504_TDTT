W6A19

```python
def vi_tri_max(lst):
    if not lst:
        return []
    m = max(lst)
    return [i for i in range(len(lst)) if lst[i] == m]
lst = list(map(int, input().split()))
res = vi_tri_max(lst)
print(*res)
```

```python
def first(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
lst = list(map(int, input().split()))
k = int(input())
res = first(lst, k)
print(res)
```

```python
def zero_to_front(lst):
    zeros = [x for x in lst if x == 0]
    others = [x for x in lst if x != 0]
    return zeros + others
lst = list(map(int, input().split()))
res = zero_to_front(lst)
print(*res)
```

```python
def tong_cuc_dai(lst):
    s = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            s += lst[i]
    return s
n = list(map(int, input().split()))
a = tong_cuc_dai(n)
print(a)
```

```python
def min(lst, x):
    kq = []
    for i in lst:
        if i < x:
            kq.append(i)
    if not kq:
        return -1
    return kq

lst = list(map(int, input().split()))
x = int(input())
d = min(lst, x)
print(*d) 
```

```python
def ktra_day(lst):
    if len(lst) < 2:
        return "vô danh"

    tang = True
    giam = True

    for i in range(1, len(lst)):
        if lst[i] <= lst[i-1]:
            tang = False
        if lst[i] >= lst[i-1]:
            giam = False

    if tang:
        return "tăng thực sự"
    if giam:
        return "giảm thực sự"
    return "vô danh"
lst = list(map(int, input().split()))
res = ktra_day(lst)
print(res)
```
